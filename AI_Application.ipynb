```python
#task3
import pandas as pd
data1 = {'key': [1, 2, 3, 4], 'value1': ['A', 'B', 'C', 'D']}
data2 = {'key': [2, 3, 4, 5], 'value2': ['W', 'X', 'Y', 'Z']}

dataset1 = pd.DataFrame(data1)
dataset2 = pd.DataFrame(data2)
inner_join = pd.merge(dataset1, dataset2, on='key', how='inner')
left_join = pd.merge(dataset1, dataset2, on='key', how='left')
right_join = pd.merge(dataset1, dataset2, on='key', how='right')
outer_join = pd.merge(dataset1, dataset2, on='key', how='outer')
print("Inner Join:")
print(inner_join)
print("\nLeft Join:")
print(left_join)
print("\nRight Join:")
print(right_join)
print("\nOuter Join:")
print(outer_join)
```

```python
#Task4
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'A': np.random.rand(50),
    'B': np.random.rand(50),
    'C': np.random.randn(50),
    'D': np.random.randn(50),
}

df = pd.DataFrame(data)

# Bar Plot
df.plot(kind='bar', figsize=(10, 6), title='Bar Plot', xlabel='Index', ylabel='Values')
plt.show()

# Line Plot
df.plot(kind='line', figsize=(10, 6), title='Line Plot', xlabel='Index', ylabel='Values')
plt.show()

# Scatter Plot
df.plot(kind='scatter', x='A', y='B', s=50, title='Scatter Plot', xlabel='A', ylabel='B')
plt.show()
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Histograms
df.plot(kind='hist', bins=20, alpha=0.5, figsize=(12, 6), title='Histograms')
plt.show()

# Box Plots
df.plot(kind='box', figsize=(10, 6), title='Box Plots')
plt.show()
```

```python
#Task5
import numpy as np

# 1. Create a NumPy array 'arr' with values from 1 to 10.
arr = np.arange(1, 11)

# 2. Create another NumPy array 'arr2' with values from 11 to 20.
arr2 = np.arange(11, 21)

# 3. Add, subtract, multiply, and divide 'arr' and 'arr2'. Print the results.
add_result = arr + arr2
sub_result = arr - arr2
mul_result = arr * arr2
div_result = arr / arr2

print("Addition:", add_result)
print("Subtraction:", sub_result)
print("Multiplication:", mul_result)
print("Division:", div_result)
```

```python
#Task6
# 1. Reshape 'arr' into a 2x5 matrix.
arr_2d = arr.reshape(2, 5)

# 2. Transpose the matrix obtained in the previous step.
transposed_arr = arr_2d.T

# 3. Flatten the transposed matrix into a 1D array.
flattened_arr = transposed_arr.flatten()

# 4. Stack 'arr' and 'arr2' vertically. Print the result.
stacked_arr = np.vstack((arr, arr2))

print("Reshaped 2x5 Matrix:", arr_2d)
print("Transposed Matrix:", transposed_arr)
print("Flattened Array:", flattened_arr)
print("Vertically Stacked Array:", stacked_arr)
```

```python
#task7
# 1. Calculate the mean, median, and standard deviation of 'arr'.
mean_arr = np.mean(arr)
median_arr = np.median(arr)
std_dev_arr = np.std(arr)

# 2. Find the maximum and minimum values in 'arr'.
max_value = np.max(arr)
min_value = np.min(arr)

# 3. Normalize 'arr' (subtract the mean and divide by the standard deviation).
normalized_arr = (arr - mean_arr) / std_dev_arr

print("Mean:", mean_arr)
print("Median:", median_arr)
print("Standard Deviation:", std_dev_arr)
print("Max Value:", max_value)
print("Min Value:", min_value)
print("Normalized Array:", normalized_arr)
```

```python
#Task 8
# 1. Create a boolean array 'bool_arr' for elements in 'arr' greater than 5.
bool_arr = arr > 5

# 2. Use 'bool_arr' to extract the elements from 'arr' that are greater than 5.
filtered_arr = arr[bool_arr]

print("Boolean Array:", bool_arr)
print("Filtered Array:", filtered_arr)
```

```python
#Task 12
# 1. Create a 2D array 'matrix' with values from 1 to 9.
matrix = np.arange(1, 10).reshape(3, 3)

# 2. Subtract the mean of each row from each element in that row.
mean_per_row = np.mean(matrix, axis=1, keepdims=True)
result_broadcasting = matrix - mean_per_row

print("Original Matrix:", matrix)
print("Result after Broadcasting:", result_broadcasting)
```
